```python
# 加载模块
import pandas as pd
import numpy as np
```

```python
# 载入CSV数据
df = pd.read_csv("IF888.csv")
```

```python
# 打印数据
df
```

```python
# 绘制收盘价
df["close"].plot()
```

```python
# 定义均线参数
fast_window = 18
slow_window = 24

# 计算每日涨跌
df["pre_close"] = df["close"].shift(1)
df["change"] = df["close"] - df["pre_close"]

# 计算移动均线
df["fast_ma"] = df["close"].rolling(window=fast_window).mean()
df["slow_ma"] = df["close"].rolling(window=slow_window).mean()
```

```python
# 绘制均线图
plot_window = 100
df["close"][:plot_window].plot()
df["fast_ma"][:plot_window].plot()
df["slow_ma"][:plot_window].plot()
```

```python
# 向量化计算均线多空信号
df["signal"] = np.where(df["fast_ma"] > df["slow_ma"], 1, -1)

# 持仓要向后平移一个周期，否则会出现未来函数
df["pos"] = df["signal"].shift(1)    # 正确
# df["pos"] = df["signal"]               # 错误

# 仓位乘以每日涨跌，得到盈亏
df["pnl"] = df["change"] * df["pos"]

# 累计求和得到总盈亏数据
df["balance"] = df["pnl"].cumsum() * 300
```

```python
# 绘制资金曲线
df["balance"].plot()
```
